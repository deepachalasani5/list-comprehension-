```python
# with function

def kamal(x):#x = user input value
    if x % 2 == 0:
        return 'Even'
    else:
        return 'Odd'
    

    
print(kamal (int(input("Enter a Number = ")) ))   

```

```python
# without comprehension

even = []
odd = []

for i in range(1,11):
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)
print('Even values = ' , even)
print('Odd values = ' , odd)
```

```python
## with list comprehesion

a = ['Even' if i % 2 ==0 else 'Odd' for i in range(1,11)]
print(a)
```

```python
even = [i for i in range(1, 11) if i  % 2 == 0]
print(even) 
```

```python
# without comprehesion
even = []
odd = []
for i in range(1,11):
    if i % 2 == 0:
        even.append(i ** 2)
    else:
        odd.append(i ** 3)
print(even)
print(odd)
```

```python
# with comprehesion
a = [i ** 2 if i % 2 == 0 else i ** 3 for i in range(1,11)]
print(a)
```

```python
## without comprehesion

a = [1,20,30,4]
b = [1,21,31,4]

for i in a: # i = 1,20,30,4
    for j in b: # j = 1 , 21 , 31 , 4
        if i == j:
            print(i)
            
    

```

```python
## with comprehension

a = [1,20,30,4]
b = [1,21,31,4]
c = [i for i in a for j in b if i == j]
print(c)
```
